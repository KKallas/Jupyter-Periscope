Setptools

Simple tools to create steps to click or type

```python
import pyperclip

def _create_new_cell(contents):
    from IPython.core.getipython import get_ipython
    shell = get_ipython()

    payload = dict(
        source='set_next_input',
        text=contents,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)

def create():
    _create_new_cell(pyperclip.paste())
    
w.create = create
```

```python
w.create()
```
